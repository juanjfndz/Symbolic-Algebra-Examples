# Fourier Transform of Gaussian Function

The Fourier Transform is a mathematical tool that allows us to convert a function from the time (or spatial) domain to the frequency domain. In other words, it decomposes a function into its constituent frequencies.

In this example, we will compute the Fourier Transform of the Gaussian function given by:

$$f(x) = e^{-x^2/2}$$

The Fourier Transform of this function is given by:

$$\mathcal{F}\{f(x)\}(k) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} e^{-ikx} e^{-x^2/2} dx$$

where $k$ is the frequency variable. We can use the `sympy` library in Python to compute this integral symbolically.

Once we have the Fourier Transform, we can plot its real and imaginary parts as a function of frequency $k$. The real part represents the cosine component of the function, while the imaginary part represents the sine component.

The Python script uses the following libraries:

- `sympy`: a Python library for symbolic mathematics
- `numpy`: a Python library for numerical computing
- `matplotlib`: a Python library for data visualization

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

## Theoretical Implementation

```python
def fourier_transform(func, var, k):
    """
    Computes the Fourier transform of a given function.

    Parameters
    ----------
    func : sympy expression
        The function to transform.
    var : sympy symbol
        The independent variable of the function.
    k : sympy symbol
        The frequency variable.

    Returns
    -------
    sympy expression
        The Fourier transform of the given function.
    """
    return 1 / sp.sqrt(2*sp.pi) * sp.integrate(f1*sp.exp(-sp.I*k*x), (x, -sp.oo, sp.oo)) 
```

```python
# Define the two functions to compare
x = sp.symbols('x')
f1 = sp.exp(-x**2/2)

# Compute the Fourier transforms of each function
k = sp.symbols('k')
F1 = fourier_transform(f1, x, k)

# Display the function and its Fourier transform
print('Function: f(x) ='); display(f1)
print('Fourier Transform: F(k) ='); display(F1)
```

## Representation

```python
# Convert the sympy expressions to numpy functions for plotting
F1_np = sp.lambdify(k, F1)

# Set up the plot
fig, ax1 = plt.subplots(figsize=(15, 6))
ax2 = ax1.twinx() # Create a second y-axis on the right side of the plot
ax1.set_xlabel('Frequency (k)', fontsize=16)
ax1.set_ylabel('Real Part', color='tab:blue', fontsize=16)
ax2.set_ylabel('Imaginary Part', color='tab:orange', fontsize=16)

# Plot the Fourier transform
k_vals = np.linspace(-5, 5, 1000)
F1_vals = F1_np(k_vals)
ax1.plot(k_vals, F1_vals.real, 
         label=r'$\mathcal{F}\{f(x)\}$', color='tab:blue')
ax2.plot(k_vals, F1_vals.imag, 
         label='', color='tab:orange', alpha=0.2)

# Add a title and legend to the plot
ax1.set_title('Fourier Transform of Gaussian Function', fontsize=20)
ax1.legend(loc='upper left', prop={'size': 16})

# Set x-axis limits
ax1.set_xlim([-5, 5])

# Set the size of the tick labels
ax1.tick_params(labelsize=14)
ax2.tick_params(labelsize=14)

# Show the plot
plt.show()
```
